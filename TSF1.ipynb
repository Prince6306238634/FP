**Importing Libraries**

```python
#importing necessary libraries/packages

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn import linear_model
import seaborn as sb
%matplotlib inline
```

**Data Collection and Pre Processing**

```python
#importing data set from drive


df = pd.read_excel('/content/drive/MyDrive/TSF/tsf.xlsx')     #for excel dataset import
print("Data Collected Successfully...")
                               #show dataframe
```

**Visualization of DataSet**

```python
#plotting the points

plt.scatter(df.Hours, df.Scores, marker='+', color='red')
plt.xlabel('Hours ->')
plt.ylabel('Scores ->')
```

```python
sb.pairplot(df,)
```

**Training the Model**

```python
#Performing Linear_Regression

reg=linear_model.LinearRegression()   #instantiating a linear regression mode

reg.fit(df[['Hours']],df.Scores)   #training the model with data
```

**Predicting data** 

```python
#Extracting Hours column to predict all its values

val1=df.iloc[:,0:1]


val1
```

```python
val2=df.iloc[:,1:2]
val2
```

```python
p= reg.predict(val1)
print("")
p
```

```python
#df1=pd.DataFrame({val2 - p })
#df1

val1['Predicted']=p
val1['Actual']=val2

val1
```

```python
#Predicting the required data

reg.predict([[9.25]])                 #Hours=9.25
```

```python
#verifing the predicted data
#m_calculation

m=reg.coef_ 
print("Slope/coefficient value, m=")
m
```

```python
#c_calculation

c=reg.intercept_ 
print("Intercept value, c=")
c
```

```python
#calucating y=mx+c
x=9.25      #prdicting hours=9.25
y=m*x+c
print("Predicted score at 9.25 hours/day study is:")
y
```

```python
#direct method to predict

print("Predicted score at 9.25 hours/day study is:")
reg.predict([[9.25]]) 
```

**Regression Line**

```python
#drawing regression straight line and final diagram
 
plt.scatter(df.Hours, df.Scores, marker='+', color='red')
plt.xlabel('Hours ->')
plt.ylabel('Scores ->')
plt.plot(df.Hours,reg.predict(df[['Hours']]),color='blue')
```
